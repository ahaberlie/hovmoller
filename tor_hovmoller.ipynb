```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.ndimage.filters import gaussian_filter as gf

#set figure size to be larger than default
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
#read csv from my github profile
torData = pd.read_csv('https://raw.githubusercontent.com/ahaberlie/hovmoller/master/torn_reports_minimal.csv',delimiter=',',header=0)

#generate a list of hours based on time of report
hour = [int(x.split(':')[0]) for x in torData['Time']]

#create a new field in the dataframe
torData['Hour'] = pd.Series(hour,index=torData.index)
```

```python
#This CSV is a minimal version of the SPC tornado report dataset at http://www.spc.noaa.gov/wcm/data/1950-2013_torn.csv

print torData.columns.values.tolist()
```

```python
#list for counting occurrence by longitude
current_hour = []

#placeholder for the hovmoller diagram
#24 hours by 70 degrees longitude
hovmoller = np.zeros((24,70))

#for every hour
for i in range(0,24):
    #get reports from current hour
    subset = torData[torData['Hour'] == i] 
    
    #remove weird lats
    subset = subset[subset['LAT'] > 0]
    
    #Pick a range.. I picked 130 W to 60 W
    for j in range(-130,-60):
        range_count = None
        current_range = None
        
        #Get reports for current whole (i.e. integer) longitude (west to east)
        current_range = subset[subset['LON']>=j]
        current_range = current_range[current_range['LON']<j+1]
        
        #get count of longitudinal subset
        range_count = len(current_range)
        
        #add count to current hour
        current_hour.append(range_count)
    
    #convert longitudinal count to numpy array
    test_arr = np.array(current_hour)
    
    #assign to current hour in hovmoller list
    hovmoller[i,:] = test_arr
    
    #initialize for the next hour
    current_hour = []
```

```python
hovmoller_filtered_all = gf(hovmoller,2)
plt.imshow(hovmoller_filtered_all,interpolation='nearest',aspect='auto',cmap=plt.cm.get_cmap('coolwarm'))

#set up x-axis labels
loc,lon = plt.xticks()
lon = range(-130,-50,10)
lons = [str(x)+"W" for x in lon]
plt.xticks(loc[1:8],lons[:-1])

#set up y-axis labels
loc,time = plt.yticks()
time = [str(int(x)).zfill(2)+" CST" for x in loc]
plt.yticks(loc[1:6],time[1:6])

plt.title('Tornado count (1950-2013) - longitude vs. hour of day')

plt.colorbar()
plt.grid()
```
