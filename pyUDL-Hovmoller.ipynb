```python
from pydap.client import open_url
from pyudl.tds import TDSCatalog

import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from IPython.display import clear_output
import sys
%matplotlib inline
```

```python
def generateHovmoller(img_list,time_step,lon0,lat0,x_step,y_step):
    """ generate a hovmoller diagram for spatiotemporal analyses
    see: http://nldr.library.ucar.edu/repository/assets/osgc/OSGC-000-000-004-222.pdf
    Figure 10 for an example
    
    img_list: representing time series of a single variable (datetime,ndarray)
    time_step (datetime): amount of time to aggregate images
    lon0 (float): westernmost longitude of images (negative in North America)
    lat0 (float): northernmost latitude of images (positive in North America)
    x_step (float): horizontal resolution in longitude for each pixel
    y_step (float): vertical resolution in latitude for each pixel
    """
    
    #Get datetime entry from first row in img_list
    start_time = img_list[0][0]
    print "start time is: ", start_time
    
    #end endtime from last row in img_list
    end_time = img_list[-1][0]
    print "end time is: ", end_time
    
    #get count of y direction pixels
    y_size = img_list[0][1].shape[0]
    print "x size is: ", y_size
    
    #get count of x direction pixels
    x_size = img_list[0][1].shape[1]
    print "y size is: ", x_size
    
    #if we are aggregating temporal slices of images, determine the ratio of total images
    #to amount of images in each slice to determine the length of the hovmoller diagram
    length = (end_time-start_time).total_seconds() / time_step.total_seconds()
    step_end_time = start_time + time_step
    
    #create aggregate image placeholder for the US
    runningSum = np.zeros((y_size,x_size))
    
    #create hovmoller placeholder
    hovmollerMap = np.zeros((length,x_size))
    
    #keep track of the current hovmoller slie
    slice_count = 0
    
    #read through each image
    for i in range(len(img_list)):
       
       #get current image
       I = img_list[i][1]
        
       #get current time to determine when slices occurr
       currentTime = img_list[i][0]
        
       #if current time is after end time end the loop
       #this shouldn't happen, but we'll check anyway!
       if currentTime > end_time:
           break
      
       #if image is in current hovmoller slice, add image to 2D US aggregate
       #else, add image to current slice, sum image over the x-axis
       #change time constraints to move to next slice, zero out 2D US aggregate 
       #for next hovmoller slice
       if currentTime < step_end_time:
           runningSum += I
       else:
           runningSum += I
           longitude_sum = np.sum(runningSum>0,axis=0)
           hovmollerMap[slice_count,:] = longitude_sum
           slice_count += 1
           start_time += time_step
           step_end_time = start_time + time_step
           runningSum = np.zeros((y_size,x_size))
                    
    return hovmollerMap
```

```python
#change start time to a day ago so we have a larger pool of images
today = dt.datetime.utcnow() - dt.timedelta(days=1)

#url string used to access composite images from unidata
url = "http://atm.ucar.edu/thredds/catalog/nexrad/composite/gini/n0r/1km/{}/catalog.xml".format(today.strftime("%Y%m%d"))

#call TDSCatalog constructor, build object with information from url string
cat = TDSCatalog(url)

#get list of image names (.gini)
names = cat.datasets.keys()

#sort images sequentially
names.sort()

#regular python list used for building (datetime,image) entries
build_list = []

#used to display progress bar in ipython
#can remove these and the loading images printline if desired
progress = 0.
img_count = 100

#loop through each of the available images,
#starting at index length - img_count
for name in names[-img_count:]:
    
    #build dataset object
    latestDs = cat.datasets[name]
    dataset = open_url(latestDs.accessUrls['OPENDAP'])
    
    #get current image time in milliseconds, convert to datetime
    time = dt.datetime.fromtimestamp(dataset["time"][:][0] / 1000.)
    
    #get current image data, threshold by 40 dBZ
    data = np.squeeze(dataset["Reflectivity"]) >= 40
    
    #add datetime and image to list
    build_list.append((time,data))
    
    #this is to show the progress of the image download and list building
    progress += 1.
    clear_output()
    print "loading images...", "{0:.1f}".format(progress/img_count * 100),"%"
    sys.stdout.flush()

#convert to numpy array
img_list = np.array(build_list)

#destroy list.. it can get pretty big
build_list = None
```

```python
#bounding box provided by unidata
llcrnrlon = -120.02283
llcrnrlat = 23.0132
urcrnrlon = -63.5105
urcrnrlat = 47.30635

#find the pixel size in x and y direction in lat/lon
lon_diff = -(llcrnrlon - urcrnrlon)
lat_diff = (urcrnrlat - llcrnrlat)
x_size = lon_diff / float(data.shape[1])
y_size = lat_diff / float(data.shape[0])
 
#if you stop the process early, the last datetime,image may be corrupted
#use this line to ignore it
#img_list = img_list[:-1]
```

```python

#find size of x and y pixels
x_length = data.shape[1]
y_length = data.shape[0]

#set the temporal resolution of hovmoller diagram
t_step = dt.timedelta(minutes=5)

#generate hovmoller data 
print "generating hovmoller data..."
sys.stdout.flush()
hov = generateHovmoller(img_list,t_step,llcrnrlon,urcrnrlat,x_size,y_size)

#plot the percent of grids that met the threshold on the latitudinal line
plt.imshow(100*hov/y_length,interpolation='nearest',aspect='auto',cmap=plt.cm.get_cmap('coolwarm'))

#set tick interval on plot
x_axis_interval = 750
y_axis_interval = 5

#generate x-tick locations
x_tick_locations = range(0,x_length,x_axis_interval)

#using x-tick locations, calculate longitudes for labelling
lons = [(llcrnrlon + loc*x_size) for loc in x_tick_locations]
lons_labelled = ["{0:.1f}".format(-x)+"W" for x in lons]

#display the x-axis ticks
plt.xticks(x_tick_locations,lons_labelled)

#for change of 1 in y-direction, how much does the time increase or decrease?
y_axis_step_conversion = int((t_step.total_seconds() / dt.timedelta(minutes=5).total_seconds()))
locs = range(0,len(hov),y_axis_interval)
timestamp = []

#build y-axis timestamp list
for i in range(0,len(img_list),y_axis_step_conversion*y_interval):
    currentTime = str(img_list[i,0].hour).zfill(2) + ":" + str(img_list[i,0].minute).zfill(2)
    timestamp.append(currentTime)

#display y-axis ticks
plt.yticks(locs[1:-1],timestamp[:-1])

#plot the title of the graph
plt.title("Latitudinal occurrence of convection")

#plot the colorbar showing count/percent/etc.
cbar = plt.colorbar()

#plot the colorbar label
cbar.set_label("% latitudinal grids >= 40 dbz")

#plot the x-axis label
plt.xlabel("Longitude")

#plot the y-axis label
plt.ylabel("Time (UTC)")

#plot grids to make interpretation easier
plt.grid()
```
